```python
import matplotlib.pyplot as plt

labels = ['Nokia', 'Samsung', 'Apple', 'Lumia']
values = [20, 30, 45, 10]
colors = ['yellow', 'green', 'red', 'blue']

figure = plt.figure()
axes = figure.add_subplot()
axes.pie(values,
         labels=labels,
         colors=colors,
         explode=[0.3, 0, 0, 0],
         shadow=True,
         autopct="%2.1f%%",
         startangle=180)
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ['milktea','blacktea', 'greentea','coffee']
values = [50, 25, 30, 40]
colors = ['purple', 'pink', 'green', 'brown']

figure = plt.figure()
axes = figure.add_subplot()
axes.pie(values,
         labels=labels,
         colors=colors,
         explode=[0.2, 0, 0, 0],
         startangle=90,
         shadow=True,
         autopct="%1.1f%%")
plt.show()
```
