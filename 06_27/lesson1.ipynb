```python
import numpy as np
import math
import matplotlib.pyplot as plt

t = np.arange(0, 2.5, 0.1)
#print(len(t))
t1 = t * math.pi
y1 = np.sin(t1)
y2 = np.cos(t1)
print(len(y1))

fig = plt.figure()
axes = fig.add_subplot(1,1,1)
axes.plot(t, y1)
axes.plot(t, y2)
plt.show()
```
